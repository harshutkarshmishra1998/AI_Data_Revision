```python
# Frequency counter for words in paragraph

import string

text = "Coffee is more than just a morning beverage; for many, coffee is a daily ritual. People love coffee because coffee provides a quick burst of energy, but the love for coffee goes deeper than just caffeine. The aroma of freshly brewed coffee can instantly brighten a gloomy day, making coffee the ultimate comfort drink. Whether you prefer iced coffee, hot coffee, or a complex espresso blend, coffee remains a universally cherished beverage across the globe."
find_text = "coffee"

cleaned_text = text.lower()
cleaned_text = cleaned_text.translate(str.maketrans('', '', string.punctuation)) # x, y = character x will be replaced by y. z = characters will be deleted
cleaned_text = cleaned_text.split()

print(cleaned_text)

counter = 0

for i in cleaned_text:
    if find_text == i:
        counter+=1

print(counter)
```

```python
# Remove duplicates from list while preserving order

orders = [
    "apple", "banana", "apple", "orange", "banana", 
    "grape", "apple", "Apple", "orange", "mango", 
    "banana", "grape", "MANGO", "apple", "kiwi"
]

orders = [fruits.lower() for fruits in orders]

orders = list(dict.fromkeys(orders)) # Returns unique without tampering ordering

print(orders)

orders = list(set(orders)) # Returns unique but it sorts them too

print(orders)
    
```

```python
# Create dictionary maintaining frequency

orders = [
    "apple", "banana", "apple", "orange", "banana", 
    "grape", "apple", "Apple", "orange", "mango", 
    "banana", "grape", "MANGO", "apple", "kiwi"
]

dictionary = {}

for i in orders:
    i = i.lower()
    if i in dictionary:
        dictionary[i]+=1
    else:
        dictionary[i]=1

print(dictionary)

top_k = 3
for k, v in dictionary.items():
    if top_k == 0:
        break
    print(k, ":", v)
    top_k-=1

print(set(orders))
```

```python
# Reverse string without slicing

word = "Utkarsh"
word_list = list(word)
length = len(word_list)

for i in range(0, int(length/2)):
    temp = word_list[i]
    word_list[i] = word_list[length-i-1]
    word_list[length-i-1] = temp

print("".join(word_list))
```

```python
# Flatten nested list

nested_list = [1, [2, 3], [[4, 5], 6], [7, [8, [9]]]]

open_list = str(nested_list)
print(open_list)

open_list = open_list.translate(str.maketrans(' ', ' ', string.punctuation))
print(open_list)

open_list = open_list.split()
print(open_list)

open_list = [int(num) for num in open_list]
print(open_list)
```

```python
# Flatten nested list

nested_list = [1, [2, 3], [[4, 5], 6], [7, [8, [9]]]]

def flatten_list(nested_list):
    flat = []
    for item in nested_list:
        if isinstance(item, list):
            flat.extend(flatten_list(item))
        else:
            flat.append(item)
    return flat

print(flatten_list(nested_list))
```

```python
# Two Sum problem

nums = [-5, 12, -1, 0, 4, -8, 7, 3, -3]
target = -2

nums.sort()

left = 0
right = len(nums) - 1

while left < right:
    sum = nums[left] + nums[right]
    if (sum == target):
        break
    elif (sum > target):
        right-=1
    else:
        left+=1

if (nums[left] + nums[right] != target):
    print("Two sum failed")
else:
    print(nums[left], "+", nums[right], "=", target)
```

```python
# Find longest substring without repeating characters

s = "dvdf"

s_list = list(s)

sub = ""
dictionary = {}

for i in range(0, len(s_list)):
    sub_temp = ""
    for j in range(i, len(s_list)):
        if s_list[j] not in dictionary:
            dictionary[s_list[j]] = 1
        else:
            break

    for k, v in dictionary.items():
        sub_temp = sub_temp + k

    if (len(sub) < len(sub_temp)):
        sub = sub_temp
    sub_temp = ""
    dictionary.clear()

print(sub)
```
